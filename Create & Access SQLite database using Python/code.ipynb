```python
#load sqlite3
import sqlite3
```

```python
# Connecting to sqlite
# connection object
conn = sqlite3.connect('INSTRUCTOR.db')
```

```python
# cursor object
cursor_obj = conn.cursor()
```

```python
#Before creating a table, let's first check if the table already exists or not. To drop the table from a database, use the DROP query. A cursor is an object that helps execute the query and fetch the records from the database.

cursor_obj.execute("DROP TABLE IF EXISTS INSTRUCTOR")
```

```python
# Creating table
table = "CREATE TABLE IF NOT EXISTS INSTRUCTOR(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CCODE CHAR(2))"

cursor_obj.execute(table)
print("table is ready")
```

```python
cursor_obj.execute("INSERT INTO INSTRUCTOR VALUES (1, 'Rav', 'Ahuja', 'TORONTO', 'CA')")
```

```python
cursor_obj.execute('''insert into INSTRUCTOR values (2, 'Raul', 'Chong', 'Markham', 'CA'), (3, 'Hima', 'Vasudevan', 'Chicago', 'US')''')
```

# Query data in the table

```python
statement = "SELECT * FROM INSTRUCTOR"

cursor_obj.execute(statement)

print("all the data")
output = cursor_obj.fetchall()

for row in output:
    print(row)
```

```python
## Fetch few rows from the table

statement = "SELECT * FROM INSTRUCTOR"

cursor_obj.execute(statement)

output = cursor_obj.fetchmany(2)
for row in output:
    print(row)
```

```python
# Fetch only FNAME from the table

statement = "SELECT FNAME FROM INSTRUCTOR"

cursor_obj.execute(statement)

output = cursor_obj.fetchall()

for row in output:
    print(row)
```

```python
#execute an update statement that changes the Rav's CITY to MOOSETOWN
query_update = 'UPDATE INSTRUCTOR SET CITY= "Moosetown" WHERE FNAME="Rav"'
cursor_obj.execute(query_update)

statement = 'SELECT * FROM INSTRUCTOR'
cursor_obj.execute(statement)
output1= cursor_obj.fetchall()

for row in output1:
    print(row)
```
