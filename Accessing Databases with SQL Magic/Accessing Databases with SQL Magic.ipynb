### Assigning the Results of Queries to Python Variables

```python
#Run this before your query to Increase SQL magic display limit
%config SqlMagic.displaylimit = 20

test_score_distribution = %sql SELECT test_score as "Test Score", COUNT(test_score) as "Frequency" FROM INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY test_score
test_score_distribution
```

### Converting Query Results to DataFrames

```python
dataframe = test_score_distribution.DataFrame()

dataframe
import seaborn as sns
plot = sns.barplot(x= "Test Score", y= "Frequency", data=dataframe)
```
